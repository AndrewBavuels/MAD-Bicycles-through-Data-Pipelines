1. Tenemos **12 columnas.** Solo necesitamos **4** de ellas:

    - **stationName:** Nombre de la estación de biciPARK
    - **address:** Dirección de la estación de biciPARK
    - **zip_code:** Código postal
    - **geometry.coordinates:** Coordenadas (longitud y latitud) de la ubicación de biciPARK

Igual que en el Dataframe BiciMAD, voy a revisar que tipo de dato son algunos elementos de **geometry.coordinates**:

```python
print(biciparks_df['geometry.coordinates'][0])
print(type(biciparks_df['geometry.coordinates'][0]))
print(type(biciparks_df['geometry.coordinates'][5]))
```

### Modificando el dataframe con las 4 columnas relevantes

```python
biciparks_df = biciparks_df[['stationName','address','zip_code','geometry.coordinates']]
```

```python
biciparks_df
```

Haré el mismo tratamiento de biciMAD en la columna **geometry.coordinates:**

```python
coordinates_df = biciparks_df['geometry.coordinates'].str.strip('[]').str.split(',', expand=True).astype('float64')
coordinates_df
```

```python
coordinates_df.columns = ['park_long', 'park_lat']
coordinates_df
```

**Concatenar los dataframes por columnas:**

```python
biciparks_df = pd.concat([biciparks_df,coordinates_df],axis=1)
biciparks_df
```

```python
# Elimino la columna geometry.coordinates
biciparks_df = biciparks_df.drop(columns=['geometry.coordinates'])
```

### Dataframe final de biciparks_df

```python
# Renombramos las columnas
biciparks_df = biciparks_df.rename(columns={"stationName": "BiciPARK station",
                          "address" : "BiciPARK address"})
biciparks_df
```
